### Load tensorflow

```python
import tensorflow as tf
```

```python
tf.keras.__version__
```

### Collect Data

```python
(features, actual_prices),_ = tf.keras.datasets.boston_housing.load_data(test_split=0)
```

```python
print('Number of examples: ', features.shape[0])
print('Number of features for each example: ', features.shape[1])
print('Shape of actual prices data: ', actual_prices.shape)
```

### Building the graph in Keras

```python
#Initialize Sequential Graph (model)
model = tf.keras.models.Sequential()

#Normalize input data
model.add(tf.keras.layers.BatchNormalization(input_shape=(13,)))

#Add a hidden layer
model.add(tf.keras.layers.Dense(6,activation='tanh'))

#Add Dense layer for prediction - Keras declares weights and bias automatically
model.add(tf.keras.layers.Dense(1))

#Compile the model - add Loss and Gradient Descent optimizer
model.compile(optimizer='sgd', loss='mse')
```

### Check model Summary

```python
model.summary()
```

### Execute the Graph

- Use batch size = 40
- use epochs = 100
- use 20% of training data as validation data 

```python
history=model.fit(features, actual_prices,batch_size=40, epochs=100,validation_split=.2)
```

### Plot the loss of validation and train data

```python
# Plot Loss in Train vs Test data
import pandas as pd
historydf = pd.DataFrame(history.history, index=history.epoch)
import plotly.graph_objs as go#visualization
train_loss = {
  "mode": "lines+markers", 
  "name": "Train Epoch Loss", 
  "type": "scatter", 
  "x":history.epoch, 
  "y":  history.history['loss']
}
test_loss = {
  "mode": "lines+markers", 
  "name": "Test Epoch Loss", 
  "type": "scatter", 
  "x":history.epoch, 
  "y":  history.history['val_loss']
}

data = [train_loss,test_loss]
layout = {
  "title": "Training vs Testing Data MAE per Epoch", 
  "xaxis": {"title": "Epoch"}, 
  "yaxis": {"title": "MAE"}
}
fig = go.Figure(data=data, layout=layout)
fig.show()
```
